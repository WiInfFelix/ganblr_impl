```python
from ganblr.utils import get_demo_data
from ganblr.models import GANBLR, GANBLRPP
```

```python
from pomegranate.bayesian_network import BayesianNetwork
import pandas as pd
import numpy as np
```

```python
dataset = get_demo_data("adult")
```

```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
from sklearn.model_selection import train_test_split

# select all but last column
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=200)
```

```python
model = GANBLR()
```

```python
model.fit(X_train, y_train, k=1, epochs = 10, batch_size = 64)
```

```python
synth_data = model.sample()
```

```python
# convert synth_data to dataframe
synth_data = pd.DataFrame(synth_data)
```

```python
# use enc to decode features and le to decode labels
synth_X = synth_data.iloc[:, :-1]
synth_y = synth_data.iloc[:, -1]
```

```python
# create random forest classifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

# fit random forest classifier on real data
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# get accuracy score on real data
real_score = accuracy_score(y_test, rf.predict(X_test))
print(f"Accuracy score on real data: {real_score}")
```

```python
from sklearn.preprocessing import LabelEncoder

# create label encoder
le = LabelEncoder()

# fit label encoder on synthetic labels
le.fit(synth_y)

# transform synthetic labels to integers
synth_y_encoded = le.transform(synth_y)

# create random forest classifier
rf_synth = RandomForestClassifier()

# fit random forest classifier on synthetic data
rf_synth.fit(synth_X, synth_y_encoded)
```

```python
# get accuracy score on synthetic data
synth_score = accuracy_score(y_test, rf_synth.predict(X_test))
print(f"Accuracy score on synthetic data: {synth_score}")
```
